```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df['sex']=df['sex'].fillna(df['sex'].mode()[0])
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['island']=le.fit_transform(df['island'])
df
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df["body_mass_g"]=ss.fit_transform(df[["body_mass_g"]])
df
#
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df["flipper_length_mm"]=ss.fit_transform(df[["flipper_length_mm"]])
df
#
```

```python
y=df['species']
x=df.drop(['species'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
lr_model.fit(x_train,y_train)
```

```python
y_pred=lr_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(y_test,y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
```

```python
pred_dt=dt_model.predict(x_test)
pred_dt
```

```python
print(accuracy_score(y_test,pred_dt))
confusion_matrix(y_test,pred_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier()
rf_model.fit(x_train,y_train)
y_pred=rf_model.predict(x_test)
print(accuracy_score(y_test,y_pred))
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k={}
neighbors=np.arange(3,20)
for k in neighbors:
  knn=KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_train,y_train)
  y_pred=knn.predict(x_test)
  acc=accuracy_score(y_test,y_pred)
  metric_k[k]=acc
  metric_k
```

```python
metric_k
```

```python
plt.plot(metric_k.keys(),metric_k.values(),marker="*")
```

```python
knn_model=KNeighborsClassifier(n_neighbors=20)
knn_model.fit(x_train,y_train)
y_pred=knn_model.predict(x_test)
acc=accuracy_score(y_test,y_pred)
acc
```

```python
confusion_matrix(y_test,y_pred)
```

```python
numerical_cols=pg.select_dtypes(include=np.number).columns
for col in numerical_cols:
  plt.figure(figsize=(8,6))
  sns.boxplot(y=pg[col])
  plt.title(f'box plot of {col}')
  plt.show()
```

```python
pg['cut'].unique()
```

```python
pg['color'].unique()
```

```python
pg['clarity'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
pg=pd.get_dummies(pg,columns=['cut','color','clarity'],drop_first=True)
pg
```

```python
bool_col=pg.select_dtypes(include='bool').columns
pg[bool_col]=pg[bool_col].astype(int)
pg
```

```python
from scipy.stats.mstats import winsorize
for col in [ 'depth', 'table']:
  pg[col] = winsorize(pg[col], limits=(0.05, 0.05))
print("Winsorization applied to 'depth', and 'table' columns.")
```

```python
pg
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
pg[['x','depth']]=scaler.fit_transform(pg[['x','depth']])
```

```python
y=pg['price']
y
```

```python
x=pg.drop(columns=['price'])
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1=train_test_split(x,y,test_size=25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train1,y_train1)
```

```python
y_pred1=lr.predict(x_test1)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(f'mae:{mean_absolute_error(y_test1,y_pred1)}')
print(f'mse:{mean_squared_error(y_test1,y_pred1)}')
print(f'r2_score:{r2_score(y_test1,y_pred1)}')
```
